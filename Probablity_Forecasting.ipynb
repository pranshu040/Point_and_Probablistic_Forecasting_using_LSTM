```python
import numpy as np
import pandas as pd
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
data['date_time'] = pd.to_datetime(data['date_time'])
data = data.sort_values(by='date_time')
y = data['WVHT'].values
```

```python
# Define a function to prepare sequences
def prepare_sequences(data, seq_length):
    sequences = []
    targets = []
    for i in range(len(data) - seq_length):
        sequences.append(data[i:i+seq_length])
        targets.append(data[i+seq_length])
    return np.array(sequences), np.array(targets)
```

```python
# Prepare sequences and targets
sequence_length = 100  # Adjust as needed
x_seq, y_targets = prepare_sequences(y, sequence_length)
```

```python
# Split the data into train and test sets
split_ratio = 0.8  # Adjust as needed
split_idx = int(split_ratio * len(x_seq))
x_train, x_test = x_seq[:split_idx], x_seq[split_idx:]
y_train, y_test = y_targets[:split_idx], y_targets[split_idx:]
```

```python
# Define quantile loss function using TensorFlow operations
def quantile_loss(q, y_true, y_pred):
    error = y_true - y_pred
    return tf.reduce_mean(tf.maximum(q * error, (q - 1) * error))
```

```python
# Import the Dropout layer from TensorFlow
from tensorflow.keras.layers import Dropout
```

```python
# Define and train the LSTM model with quantile loss
def train_model_with_quantile_loss(q):
    model = Sequential([
        LSTM(64, activation='relu', input_shape=(sequence_length, 1)),
        Dense(1)
    ])

    model.compile(optimizer=Adam(learning_rate=0.001), loss=lambda y_true, y_pred: quantile_loss(q, y_true, y_pred))
    model.fit(x_train, y_train, epochs=50, batch_size=32, verbose=1)
    return model
```

```python
#Train models with 20% and 80% quantiles
q_10 = 0.10  # 10% quantile
model_10 = train_model_with_quantile_loss(q_10)
```

```python
q_90 = 0.90  # 90% quantile
model_90 = train_model_with_quantile_loss(q_90)
```

```python
# Make predictions
y_pred_10 = model_10.predict(x_test)
y_pred_90 = model_90.predict(x_test)
```

```python
# Plot quantile predictions along with actual values
plt.plot(data['date_time'][split_idx+sequence_length:split_idx+sequence_length+len(y_test)], y[split_idx+sequence_length:split_idx+sequence_length+len(y_test)], label="Actual")
plt.plot(data['date_time'][split_idx+sequence_length:split_idx+sequence_length+len(y_test)], y_pred_10, label="10% Prediction")
plt.plot(data['date_time'][split_idx+sequence_length:split_idx+sequence_length+len(y_test)], y_pred_90, label="90% Prediction")
plt.xlabel("Date Time")
plt.ylabel("WVHT")
plt.legend()
plt.show()                                                                                                                                                                                                  # Calculate the number of points between 95% and 5% percentiles
num_points_between_percentiles = np.sum((y_pred_10 <= y_test) & (y_test <= y_pred_90))
```

```python
import numpy as np

def calculate_picp_bootstrap(y_true, y_pred, num_bootstraps=1000, alpha=0.1, beta=0.9):
    num_within_interval_10 = 0
    num_within_interval_90 = 0

    for _ in range(num_bootstraps):
        indices = np.random.choice(len(y_true), size=len(y_true), replace=True)
        y_lower_bootstrap_10 = np.percentile(y_pred[indices], alpha * 100)
        y_upper_bootstrap_10 = np.percentile(y_pred[indices], beta * 100)
        y_lower_bootstrap_90 = np.percentile(y_pred[indices], (1 - beta) * 100)
        y_upper_bootstrap_90 = np.percentile(y_pred[indices], (1 - alpha) * 100)

        within_interval_10 = (y_lower_bootstrap_10 <= y_true) & (y_true <= y_upper_bootstrap_10)
        within_interval_90 = (y_lower_bootstrap_90 <= y_true) & (y_true <= y_upper_bootstrap_90)

        num_within_interval_10 += np.mean(within_interval_10)
        num_within_interval_90 += np.mean(within_interval_90)

    picp_10 = num_within_interval_10 / num_bootstraps
    picp_90 = num_within_interval_90 / num_bootstraps
    return picp_10, picp_90

# Assuming you have y_test and y_pred_10 as before
picp_10, picp_90 = calculate_picp_bootstrap(y_test, y_pred_10)

print("PICP (10%):", picp_10)
print("PICP (90%):", picp_90)
```
